# PUNJAB BUS ROUTE LINKS

```python
driver_P=webdriver.Chrome()

#load the webpage

driver_P.get("https://www.redbus.in/online-booking/pepsu-punjab?utm_source=rtchometile")

time.sleep(3)

driver_P.maximize_window()
#retrive bus links and route
wait = WebDriverWait(driver_P, 20)
def punjab_link_route(path):   
    LINKS_PUNJAB=[]
    ROUTE_PUNJAB=[]
    
    for i in range(1,11):
        paths=driver_P.find_elements(By.XPATH,path)
        # retrive the route links 
        for links in paths:
            d = links.get_attribute("href")
            LINKS_PUNJAB.append(d)
            
        # retrive names of the routes
        for route in paths:
            ROUTE_PUNJAB.append(route.text)
            
        try:
            # Wait for the pagination element to be present
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')
            time.sleep(4)
            next_button.click()
            
        except NoSuchElementException:
            
            break
            
    return LINKS_PUNJAB,ROUTE_PUNJAB

LINKS_PUNJAB,ROUTE_PUNJAB=punjab_link_route("//a[@class='route']")

df_P=pd.DataFrame({"Route_name":ROUTE_PUNJAB,"Route_link":LINKS_PUNJAB})

print(df_P)

# change dataframe to csv

path=r"D:\Redbus\df_P.csv"
df_P.to_csv(path,index=False)
```
